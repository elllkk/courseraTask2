```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate, linalg
```

```python
x = np.arange(1, 16, 1)
y = np.sin(x/5) * np.exp(x/10) + 5*np.exp(-x/2)
#linear
b=np.array([float(y[0]),float(y[14])])
b=b.reshape((2,1))

A=np.array([[1,1],[1,15]])
A
```

```python
w = linalg.solve(A,b)
plt.plot(x,y)
plt.plot([1,15], b)
plt.show()
```

```python
#parabola
b = np.array([float(y[0]),float(y[7]), float(y[14])])
b = b.reshape(3,1)
A = np.array([[1,1,1],[1,8,8**2],[1,15,15**2]])
w = linalg.solve(A, b)
plt.plot(x,y)
plt.plot([1,8,15], b)
plt.show()
```

```python
b = np.array([float(y[0]),float(y[3]),float(y[9]), float(y[14])])
b = b.reshape(4,1)
A = np.array([[1,1,1,1],[1,4,4**2,4**3],[1,10,10**2,10**3],[1,15,15**2,15**3]])
w = linalg.solve(A, b)
plt.plot(x,y)
plt.plot([1,4,10,15], b)
plt.show()
print(w)
```

```python
with open('submission-2.txt','w') as file_to_write:
    str_to_write= str(w[0,0])+' '+str(w[1,0])+' '+str(w[2,0])+' '+str(w[3,0])
    file_to_write.write(str_to_write)
```
